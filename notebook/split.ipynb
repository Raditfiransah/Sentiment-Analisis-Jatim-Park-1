```python
import pandas as pd

# misal df udah ada dan kolom rating = 'Rating'
df = pd.read_csv('../data/Jatim-Park-1-1000.csv')

# ambil semua data rating 1-4
low_ratings = df[df['rating'] < 5]

# hitung berapa banyak data rating 1-4
count_low = len(low_ratings)

# hitung sisa slot untuk mencapai total 500
remaining = 500 - count_low

# ambil data rating 5 secukupnya
high_ratings = df[df['rating'] == 5].sample(n=remaining, random_state=42)

# gabungkan semuanya
final_df = pd.concat([low_ratings, high_ratings])

# kocok biar random
final_df = final_df.sample(frac=1, random_state=42).reset_index(drop=True)

print(final_df['rating'].value_counts())
print(f"Total data: {len(final_df)}")
```
